```python
def check(lines):
    lines = lines.reset_index(drop=True)
    val = []
    everything = []
    err = 0
    days = lines['Date'].unique()
    for d in range(len(days)):
        print(days[d])
        current = days[d]
        avail = data.loc[data['Date']< current]
        for i in range(len(lines)):

            date = lines['Date'][i]
            if date == current:
                if i%20 == 0:
                    print(i)
                player = lines['Player'][i]

                cat = lines['Stat'][i]
                line = lines['Line'][i]
                opp = lines['Team'][i]
                game  = get_game(player,date)
                expect = adjust(player,cat,opp,stats=avail)
                expect_nokm = adjust_raw(player,cat,opp,stats=avail)

                if game.empty:
                    print('err',player)
                    err += 1
                else:
                    x = get_stat(player,cat,game)
                    pred_diff = expect-line
                    diff_real = x-line
                    if np.sign(pred_diff) == np.sign(diff_real):
                        val=1
                    else:
                        val=0
                    season = float(print_prob(player,line,cat,stats=avail,prnt=False))
                    l_10 = float(print_prob(player,line,cat,games=10,stats=avail,prnt=False))
                    everything.append((player,opp,cat,expect,expect_nokm,line,x[0],l_10,season))
            
    x = np.array(everything)
    df = pd.DataFrame(x,columns=['Player','opp','Stat','pred','pred_nokm','line','actual','l_10','season'])
    numeric = ['pred','pred_nokm','line','actual','l_10','season']
    df[numeric]=df[numeric].astype(float)
    df['Prob']= 1- poisson.cdf(mu=df['pred'],k=df['line'])
    return df
```

```python
def dynamic(guy,cat,stats=None,series=None,alpha=None):
    if stats is None:
        stats = data
    if series is None:
        series = get_stat(guy,cat,stats)
        series = np.array(series)
    if alpha==None:
        alpha = 0.2
    
    pred = [series[0]]
    mov = [series[0]]
    for i in range(1,3):
            predict = series[0:i].mean()
            pred.append(predict)
            mov.append(predict)
    for i in range(3,len(series)):
        prev = mov[i-1]
        current = series[i]*(alpha) + mov[i-1]*(1-(alpha))
        mov.append(current)
    return(mov)
```

```python
def find_dynamic(guy,cat):
    series = get_stat(guy,cat)
    series = np.array(series)
    
    idx = []
    for alpha in range(1,20):
        mov = dynamic(guy,cat,alpha=alpha/50)
        result = (((mov[10:-1]-series[11:])**2).sum())
        idx.append(result)
    return idx
```

```python
def pt_corr(player_a_name, player_b_name, team):
    team_games = data.loc[data.Tm==team]
    # extract two columns of interest (i.e. points scored by the two players)
    player_a_dates = set(team_games[team_games['Player'] == player_a_name]['Date'])
    player_b_dates = set(team_games[team_games['Player'] == player_b_name]['Date'])
    common_dates = player_a_dates.intersection(player_b_dates)
    stats = team_games[team_games['Date'].isin(common_dates)]
    
    player_a_points = stats[stats['Player'] == player_a_name]['PTS'].reset_index(drop=True)
    player_b_points = stats[stats['Player'] == player_b_name]['PTS'].reset_index(drop=True)

    # calculate the correlation coefficient between the two columns
    corr_coeff = player_a_points.corr(player_b_points)
    # return the correlation coefficient
    return corr_coeff
```
