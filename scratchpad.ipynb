```python
unabated = pd.read_csv('Lines/unabated_2023_10_6.csv',index_col=0)
unabated.columns = [x.lower() for x in unabated.columns]
```

```python
pp_lines = pd.read_csv('Lines/full_lines_2023_10_6',index_col=0)
pp_lines.columns = [x.lower() for x in pp_lines.columns]
```

```python
result = pp_lines.merge(unabated,on=['player','stat'],how='inner',suffixes=['_pp','_unabated']).sort_values(by='over_prob')
result.columns
```

```python
result = result[['player', 'line_pp', 'stat', 'line_unabated', 'event_time', 'opp', 'count',
       'over_prob', 'under_prob', 'pred']]
```

```python
result['diff'] = result['pred'] - result['line_pp']
from scipy.stats import poisson
```

```python
result['poiss'] = 1- poisson.cdf(mu=result['pred'],k=result['line_pp'])
result.sort_values(by='diff').tail(15)
```
