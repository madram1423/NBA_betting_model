```python
import urllib3
import json

import urllib
class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/5.0"

    
pp_props_url = 'https://parlayplay.io/api/v1/crossgame/search/'
opener = AppURLopener()
response = opener.open(pp_props_url)

http = urllib3.PoolManager()

url = 'https://httpbin.org/post'
myheaders = {'Content-Type': 'application/json'}
mydata = {'username': '***', 'password': '***'}
#print(dir(response))
print(response.status)

data = json.loads(response.read())
df = pd.json_normalize(data['players'])
df.head()
```

```python
df2 = pd.json_normalize(df['stats'])
pd.json_normalize(df2.iloc[0])
```

```python

lines = pd.DataFrame()
for i in range(len(df)):
    temp = pd.json_normalize(df.stats[i])
    temp['player.fullName'] = df['player.fullName'][i]
    temp['player.team.teamAbbreviation'] = df['player.team.teamAbbreviation'][i]
    temp = temp[['player.fullName','player.team.teamAbbreviation', 'challengeUnits','statValue']]
    lines = pd.concat((lines,temp))

lines = lines.rename(columns={'statValue':'Line','player.fullName':'Player',
                              'player.team.teamAbbreviation':'Team','challengeUnits':'Stat'})
#lines['Opp'] = 0
#for i in range(len(lines)):
#    if i in 
#    lines['Opp'][i]
lines.head()
```

```python
test = lines

test = test.replace('Points','PTS')
test = test.replace('P+R+A','PTS+TRB+AST')
test = test.replace('P+R','PTS+TRB')
test = test.replace('P+A','PTS+AST')
test = test.replace('R+A','TRB+AST')
test = test.replace('Rebs','TRB')
test = test.replace('Asts','AST')
test = test.replace('Rebounds','TRB')
test = test.replace('3PM','3P')
test = test.replace('Assists','AST')
test = test.replace('Turnovers','TOV')
test = test.replace('Blks','BLK')
test = test.replace('Pts','PTS')
test = test.replace('Stls','STL')
test = test.replace('Free Throws Made','FT')
test = test.replace('Blocked Shots','BLK')
test = test.replace('Steals','STL')
test = test[test.Stat != 'Fantasy Score']

lines = test
lines
```
